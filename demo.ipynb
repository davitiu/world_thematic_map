## Dataset Structure

Each visualization requires:
- A CSV file containing country-level data
- A corresponding JSON metadata file mapping country identifiers
- Each dataset type is a tuple, containing strings of paths of CSV file and corresponding JSON metadata file, this order.

This demo uses 8 datasets from **Our World in Data (OWID)** to demonstrate
robustness and flexibility.

```python
from choromap.choromap import ChoroMap
```

```python
data1=("data/lead-blood-usa-children/lead-blood-usa-children.csv", "data/lead-blood-usa-children/lead-blood-usa-children.metadata.json")
data2 = ("data/lead-concentrations-blood-children/lead-concentrations-blood-children.csv", "data/lead-concentrations-blood-children/lead-concentrations-blood-children.metadata.json")
data3 = ("data/lead-concentrations-blood-adults/lead-concentrations-blood-adults.csv", "data/lead-concentrations-blood-adults/lead-concentrations-blood-adults.metadata.json")
data4 = ("data/lead-paint-over-10000ppm/lead-paint-over-10000ppm.csv","data/lead-paint-over-10000ppm/lead-paint-over-10000ppm.metadata.json")
data5 = ("data/lead-paint-over-600ppm/lead-paint-over-600ppm.csv", "data/lead-paint-over-600ppm/lead-paint-over-600ppm.metadata.json")
data6 = ("data/lead-paint-over-90ppm/lead-paint-over-90ppm.csv", "data/lead-paint-over-90ppm/lead-paint-over-90ppm.metadata.json")
data7 = ("data/lead-petrol-ban/lead-petrol-ban.csv", "data/lead-petrol-ban/lead-petrol-ban.metadata.json")
data8 = ("data/legal-controls-lead-paint/legal-controls-lead-paint.csv", "data/legal-controls-lead-paint/legal-controls-lead-paint.metadata.json")
```

### Creating an object from the ChoroMap class that recieves multiple csv datasets and their corresponding json meta datasets. Accepts **kwargs.

```python
lead = ChoroMap(data1 = data1,
                data2 = data2,
                data3 = data3,
                data4 = data4,
                data5 = data5,
                data6 = data6,
                data7 = data7,
                data8 = data8)
```

## Example 1: Basic World Choropleth Map

This example demonstrates a standard thematic map using default settings.

```python
lead.plot_choropleth("data6")
```

## Example 2: Handling Missing Data

Real-world datasets often contain missing values.
This example highlights how the package visualizes countries with no data.


When `show_no_data=True`, countries without available data are rendered
using a dedicated placeholder value.

### Known Limitation
- Due to color contrast constraints, both the countries with data and missing-data countries may appear
  visually similar to low-value regions on the map.
- However, when hovering over a country, the corresponding value is explicitly labeled; in case of missing data,
  it is assigned as **-1**, ensuring clear interpretability.

```python
lead.plot_choropleth("data3", show_no_data=True)
```

## Example 3: Package modified colorscale

Users can specify alternative color scales to better match their data
or visual preferences. When `data_based_colorbar`  is set to True, colorscale based on statistical data will be used, or 
the users can change to alternative colorscales when `colorscale` is set accordingly.

```python
lead.plot_choropleth("data4", data_based_colorbar=True)
```

```python
lead.plot_choropleth("data5", colorscale="Magma")
```

## Data Distribution

In addition to maps, the package supports histogram visualization to
analyze value distributions.

```python
lead.plot_bar("data1")
```
